```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the graph
G = nx.Graph()

# Add edges based on the connectivity information
edges = [
    ('cell1', 'cell2'), ('cell1', 'cell8'),
    ('cell2', 'cell3'), ('cell2', 'cell9'),
    ('cell3', 'cell4'), ('cell3', 'cell8'),
    ('cell3', 'cell10'),
    ('cell4', 'cell5'), ('cell4', 'cell9'),
    ('cell4', 'cell11'),
    ('cell5', 'cell6'), ('cell5', 'cell10'),
    ('cell5', 'cell12'),
    ('cell6', 'cell7'), ('cell6', 'cell11'),
    ('cell7', 'cell12'),
    ('cell8', 'cell9'),
    ('cell9', 'cell10'),
    ('cell10', 'cell11'),
    ('cell11', 'cell12')
]

G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)  # Positions for all nodes

nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", font_size=10, font_weight="bold")
plt.title("Undirected Graph Visualization")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the connections between cells
connections = [
    ('cell1_1', 'cell1_2'), ('cell1_2', 'cell1_1'), ('cell1_2', 'cell1_3'),
    ('cell1_3', 'cell1_2'), ('cell1_3', 'cell1_4'), ('cell1_4', 'cell1_3'),
    ('cell1_4', 'cell4_1'), ('cell1_5', 'cell1_6'), ('cell1_5', 'cell4_1'),
    ('cell2_1', 'cell2_2'), ('cell2_2', 'cell2_1'), ('cell2_2', 'cell2_3'),
    ('cell2_3', 'cell2_2'), ('cell2_3', 'cell2_4'), ('cell2_4', 'cell2_3'),
    ('cell2_4', 'cell4_1'), ('cell2_5', 'cell4_1'), ('cell3_1', 'cell3_2'),
    ('cell3_2', 'cell3_1'), ('cell3_2', 'cell3_3'), ('cell3_3', 'cell3_2'),
    ('cell3_3', 'cell3_4'), ('cell3_4', 'cell3_3'), ('cell3_4', 'cell4_1'),
    ('cell3_5', 'cell3_6'), ('cell3_5', 'cell4_1'), ('cell3_6', 'cell3_5'),
    ('cell4_1', 'cell1_4'), ('cell4_1', 'cell1_5'), ('cell4_1', 'cell2_4'),
    ('cell4_1', 'cell2_5'), ('cell4_1', 'cell3_4'), ('cell4_1', 'cell3_5')
]

# Create a directed graph
G = nx.DiGraph()

# Add edges to the graph
G.add_edges_from(connections)

# Draw the graph
nx.draw(G, with_labels=True, font_weight='bold')

# Show the plot
plt.show()
```
